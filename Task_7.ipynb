```python
# 1.Smart Light Controller
class SmartLight:
  def __init__(self,lightname):
    self.lightname=lightname
    self.is_on=False
  def turn_on(self):
    self.is_on=True
  def turn_off(self):
    self.is_off=False
  def display(self):
    if self.is_on:
      print(f"{self.lightname} is ON")
    else:
      print(f"{self.lightname} is OFF")
light=SmartLight("Bedroom Light")
light.turn_on()
light.display()
```

```python
# 2.Employee ID Card System
class Employee:
  def __init__(self,empname,empid,dept):
    self.empname=empname
    self.empid=empid
    self.dept=dept
  def display(self):
    print("Employee ID Card")
    print(f"Name:{self.empname}")
    print(f"ID:{self.empid}")
    print(f"Department:{self.dept}")
empname="Rahul"
empid="EMP102"
dept="IT"
employee=Employee(empname,empid,dept)
employee.display()
```

```python
# 3.Mobile Contact Record
class Mobile:
  def __init__(self,name,phone):
    self.name=name
    self.phone=phone
  def display(self):
    print("Contact Saved")
    print(f"Name: {self.name}")
    print(f"Phone: {self.phone}")
name="Anita"
phone="9876543210"
mobile=Mobile(name,phone)
mobile.display()
```

```python
# 4.Product Price Tag Generator
class Product:
  def __init__(self,name,price):
    self.name=name
    self.price=price
  def display(self):
    print(f"Product Name: {self.name}")
    print(f"Price: ₹{self.price}")
name="Headphones"
price="2499"
product=Product(name,price)
product.display()
```

```python
#5. Movie Rating Display System
class Movie:
  def __init__(self,movie_name,rating):
    self.movie_name=movie_name
    self.rating=rating
  def display(self):
    print(f"Movie: {self.movie_name}")
    print(f"Rating: {self.rating}/5")
movie_name="Inception"
rating="4.8"
movie=Movie(movie_name,rating)
movie.display()
```

```python
# 6. Delivery Address Manager
class Delivery:
  def __init__(self,name,add):
    self.name=name
    self.add=add
  def display(self):
    print(f"Delivery Details")
    print(f"Customer: {self.name}")
    print(f"Address: {self.add}")
name="Suman"
add="Hyderabad"
delivery=Delivery(name,add)
delivery.display()
```

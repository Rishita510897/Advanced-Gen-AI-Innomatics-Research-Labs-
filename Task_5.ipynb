```python
# 1.Social Media-Post Engagement Analyzer
def analyze_engagement(likes_list):
  total_likes=0
  for like in likes_list:
    total_likes+=like
  if total_likes>=1000:
    status="Viral Post"
  else:
    status="Normal Engagement"
  print("Total Likes:",total_likes)
  print("Post Status:",status)
likes=[200,150,300,250,150]
analyze_engagement(likes)
```

```python
# 2.HealthCare-Medicine Stock Alert System
def check_medicine_stock(stock):
  if stock<10:
    status="Low Stock Alert"
  else:
    status="Stock Sufficient"
  print("Medicine Stock:",stock)
  print("Status:",status)
check_medicine_stock(6)
```

```python
# 3.Agriculture-Rainfall Adequacy Checker
def check_rainfall(rainfall_list,required_level):
  total=0
  for rain in rainfall_list:
    total+=rain
  average=total/len(rainfall_list)
  if average>=required_level:
    status="Adequate Rainfall"
  else:
    status="Inadequate Rainfall"
  print("Average Rainfall:",int(average))
  print("Rainfall Status:",status)
rainfall_data=[70,75,68,80,67]
check_rainfall(rainfall_data,70)
```

```python
# 4.Social Media-Duplicate Account Deletion
def detect_duplicate_accounts(usernames):
  unique_usernames=set(usernames)
  if len(unique_usernames)!=len(usernames):
    result="Yes"
  else:
    result="No"
  print("Duplicate Accounts Found:",result)
users=["rishi01","anita22","rishi01","john45"]
detect_duplicate_accounts(users)
```

```python
# 5.Healthcare-Appointment Eligibility Checker
def check_appointment_eligibility(age):
  if age>=18:
    status="Eligible"
  else:
    status="Not Eligible"
  print("Patient Age:",age)
  print("Eligibility Status:",status)
check_appointment_eligibility(21)
```

```python
# 6.Premium Crop Price Filter
def filter_premium_crops(prices):
  premium_crops=[]
  for price in prices:
    if price>2000:
      premium_crops.append(price)
  print("Premium Crops:",premium_crops)
crop_prices=[1500,2500,1800,3200]
filter_premium_crops(crop_prices)
```

```python
# 7.Application Health Checker
def check_application_health(errors):
  if errors==0:
    status="Healthy"
  elif errors<=5:
    status="Minor Issues"
  else:
    status="Critical Issues"
  print("Error Count:",errors)
  print("System Status:",status)
check_application_health(7)
```

```python
# 9.Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
  total_classes=len(attendance_list)
  present_count=0
  for status in attendance_list:
    if status=="P":
      present_count+=1
    attendance_percentage=(present_count/total_classes)*100
    if attendance_percentage>=75:
      eligibility="Eligible"
    else:
      eligibility="Not Eligible"
  print("Attendance Percentage:",attendance_percentage)
  print("Exam Eligibility:",eligibility)
attendance=["P","P","A","P","P"]
check_exam_eligibility(attendance)
```

```python
# 10.Smart Electricity Bill Analyzer
def calculate_electricity_bill(units):
  bill = 0
  if units <= 100:
    bill = units * 3
  elif units <= 200:
    bill = (100 * 3) + ((units - 100) * 5)
  else:
    bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
  if bill < 500:
    usage_status = "Low Usage"
  elif bill <= 1500:
    usage_status = "Moderate Usage"
  else:
    usage_status = "High Usage"
  return bill, usage_status
total_bill, status = calculate_electricity_bill(250)
print("Total Bill:", total_bill)
print("Usage Status:", status)
```

```python
# 1. Unique Words in a Sentence
sentence="python is easy and python is powerful"
words=sentence.split()
unique_words=set(words)
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
```

```python
# 2.Highest Salary from Employee Data
employees={"Ravi":75000,"Anita":68000,"Kiran":72000}
highest_employee=max(employees,key=employees.get)
print("Highest salary:",highest_employee,"-",employees[highest_employee])
```

```python
# 3. Find Maximum and Minimum Values
nums=[45, 22, 89, 10, 66]
maxi=nums[0]
mini=nums[0]
for num in nums:
  if num>maxi:
    maxi=num
  if num<mini:
    mini=num
print("List:",nums)
print("Max:",maxi)
print("Min:",mini)
```

```python
# 4. Count Products Above a Price Threshold
prices=[450, 1200, 899, 1500, 300]
cnt=0
for i in prices:
  if i>1000:
    cnt+=1
print("Products above 1000:",cnt)
```

```python
# 5. Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
present_cnt=attendance.count("p")
total_days=len(attendance)
percentage=(present_cnt/total_days)*100
print("Attendance Percentage:",percentage)
```

```python
# 6. Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_num=set(phone_numbers)
print("Unique phone numbers:",unique_num)
```

```python
# 7. Count Character Frequency
text="python"
freq={}
for char in text:
  if char in freq:
    freq[char]+=1
  else:
    freq[char]=1
print(freq)
```

```python
# 8. Convert List to Tuple
nums=[10,20,30]
t=tuple(nums)
print("Tuple:",t)
```

```python
# 9. Check if a Key Exists in a Dictionary
employees={"Ravi":75000,"Anita":68000}
if "Ravi" in employees:
  print("Employee exists")
else:
  print("Employee does not exist")
```

```python
# 10. Calculate Average Marks
marks=[70,80,65,75]
average=sum(marks)/len(marks)
print("Average Marks:",average)
```

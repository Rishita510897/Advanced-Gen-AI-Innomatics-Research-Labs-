```python
# 1. Mobile Recharge Validation System
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False
    return True
while True:
    amount = int(input("Enter recharge amount: ₹"))
    if validate_recharge(amount):
        print("Recharge successful")
        break
    else:
        print("Please try again.\n")
```

```python
# 2.Inventory Reorder Alert System
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "- Reorder Alert")
        else:
            print(product, "- Stock OK ")
inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 12
}
check_inventory(inventory)
```

```python
# 3. Student Result Processing System
def process_result(marks):
    total = 0
    for mark in marks:
        total = total + mark
    average = total / len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print("Average Marks:", sum(student_marks) / len(student_marks))
print("Result:", result)
```

```python
# 4. Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (distance * per_km_rate)
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator ")
        break
```

```python
# 5. Employee Attendance Eligibility Checker
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1
    attendance_percentage = (present_days / total_days) * 100
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_attendance(attendance)
print("Attendance Percentage:", (attendance.count("P") / len(attendance)) * 100, "%")
print("Eligibility:", result)
```

```python
# 6. Password Strength Checker
def check_password(password):
    special_chars = "@#$"
    # Check minimum length
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long"
    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    if not has_digit:
        return "Weak: Password must contain at least one digit"
    has_special = False
    for char in password:
        if char in special_chars:
            has_special = True
            break
    if not has_special:
        return "Weak: Password must contain at least one special character (@, #, $)"
    return "Strong Password "
password = input("Enter password to check: ")
result = check_password(password)
print(result)
```
